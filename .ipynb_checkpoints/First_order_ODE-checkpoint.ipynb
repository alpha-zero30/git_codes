## Euler Method 

the ode is 
$$ \dfrac{dx}{dt} = -xt $$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

t_i = 0
t_f = 15

x_in = 1

h = 0.01


def dxdt(t,x):
    return  -x*t

def euler(f,t_i,t_f,x_in,h):
    ts = np.arange(t_i,t_f,h)
    xs = np.zeros_like(ts)
    
    xs[0] = x_in
    
    for i,tn in enumerate(ts[:-1]):  # so that last term count is not needed
        xs[i+1] = xs[i] + h*dxdt(tn,xs[i])   
    return ts,xs

ts_e = euler(dxdt,t_i,t_f,x_in,h)[0]
xs_e = euler(dxdt,t_i,t_f,x_in,h)[1]





def midpoint(f,t_i,t_f,x_in,h):
    ts = np.arange(t_i,t_f,h)
    xs = np.zeros_like(ts)
    
    xs[0] = x_in
    
    for i,tn in enumerate(ts[:-1]):
        xs_half = xs[i] + 0.5*h*dxdt(tn,xs[i])
        xs[i+1] = xs[i] + h*dxdt(tn+ h/2 , xs_half)
        
    return ts,xs

ts_m = midpoint(dxdt,0,15,1,0.01)[0]
xs_m = midpoint(dxdt,0,15,1,0.01)[1]


plt.plot(ts_e,xs_e)
plt.scatter(ts_m,xs_m,c='r')
plt.tight_layout()
```

##  second order ode

second order ode
$$ \dfrac{d^2y}{dt^2} = -4\pi^2y $$

$$  \dfrac{dy}{dt} = p    \hspace{1cm} \dfrac{dp}{dt} = -4\pi^2y $$

## euler 2nd order 

```python
def slope(ys,t):
    return np.asarray([ys[1] , -4*np.pi**2*ys[0]])      # since for y the slope eq is the velocity so ys[1] and for velocity it is slope as -4pi**2 y

def euler(f,t_i,t_f,y_in,h):
    
    ts = np.arange(t_i,t_f,h)
    ys = np.zeros((len(ts),len(y_in)))
    
    ys[0]=y_in      # initializing
    
    for i,tn in enumerate(ts[:-1]):
        ys[i+1] = ys[i] + h*f(ys[i],tn)
        
    return ts,ys
    
    
ts, ys = euler(slope, 0, 15,[1,0], 0.01)


pos = ys[:,0]
vel = ys[:,1]

plt.plot(ts,pos)
plt.plot(ts,vel)


```

## Midpoint 2nd order 

```python
def slope(ys,t):
    return np.asarray([ys[1] , -4*np.pi**2*ys[0]])

def midpoint(f,t_i,t_f,y_in,h):
    
    ts = np.arange(t_i,t_f,h)
    ys = np.zeros((len(ts),len(y_in)))
    
    ys[0]=y_in
    
    for i,tn in enumerate(ts[:-1]):
        ys_half = ys[i] + 0.5*h*slope(ys[i],tn)
        ys[i+1] = ys[i] + h*f( ys_half,tn+ h/2 )
        
    return ts,ys
    
    
ts, ys = midpoint(slope, 0, 10, [1, 0], 0.01)


pos = ys[:,0]
vel = ys[:,1]

plt.plot(ts,pos,label="position")
plt.plot(ts,vel,label="velocity")
plt.ylim(-10,10)
plt.yticks(np.arange(-10,11,2))
plt.legend()
plt.show()
```

## range kutta 2nd order 

```python
def rk4(f,t_i,t_f,y_in,h):
    ts = np.arange(t_i,t_f,h)
    ys = np.zeros((len(ts),len(y_in)))
    
    ys[0] = y_in
    
    for i,tn in enumerate(ts[:-1]):
        k1 = h*f(ys[i],tn)
        k2 = h*f(ys[i] + k1/2 , tn + h/2 )
        k3 = h*f(ys[i] + k2/2 , tn + h/2 )
        k4 = h*f(ys[i] + k3 , tn + h)
        ys[i+1] = ys[i] + (k1 +2*k2 + 2*k3 + k4)/6
        
        
    return ts,ys


ts,ys = rk4(slope,0,15,[1,0],0.01)

pos = ys[:,0]
vel = ys[:,1]

plt.plot(ts,pos,label="position")
plt.plot(ts,vel,label="velocity")
plt.ylim(-10,10)
plt.yticks(np.arange(-10,11,2))
plt.legend()
plt.show()
```

# Solving ODE using verlet, velocity verlet

### verlet for position

second order ode
$$ \dfrac{d^2y}{dt^2} = -4\pi^2y $$

$$  \dfrac{dy}{dt} = p    \hspace{1cm} \dfrac{dp}{dt} = -4\pi^2y $$

```python
# verlet directly uses second derivative slope values to calculate the solns

# y1 = y0 + v0 h + 1/2 a0 h^2
#yn+1 = 2yn − yn−1 + an h**2

def a(t,y):
    return -4*(np.pi**2)*y

def verlet(f,t_i,t_f,y_in,v_in,h):
    ts = np.arange(t_i,t_f,h)
    ys = np.zeros(len(ts))
    
    ys[0] = y_in
    ys[1] = ys[0] + v_in*h + 0.5*a(ts[0],ys[0])*h**2
    
    for i,tn in enumerate(ts[1:-1]):   # so that 1st term is also used
        ys[i+2] = 2*ys[i+1] - ys[i] + a(tn,ys[i+1])*(h**2)
        
    return ts,ys

ts,ys = verlet(a,0,15,1,0,0.01)

plt.plot(ts,ys)
ys_exact = np.cos(2*np.pi*ts)

plt.plot(ts,ys_exact)
   
error = ys_exact[-1] - ys[-1]
print(error)
```

### Velocity verlet 

velocity verlet also uses same acceleration for both pos and velocity values

```python
# yn+1 = yn + vnh + anh2/2
# vn+1 = vn + h(an + an+1)/2


def a(t,y):
    return -4*(np.pi**2)*y

def vel_verlet(f,t_i,t_f,y_in,v_in,h):
    ts = np.arange(t_i,t_f,h)
    ys = np.zeros(len(ts))
    vs = np.zeros(len(ts))
    
    ys[0] = y_in
    vs[0] = v_in
    
    for i,tn in enumerate(ts[:-1]):
        ys[i+1] = ys[i] + vs[i]*h + 0.5*a(tn,ys[i])*h**2
        vs[i+1] = vs[i] + 0.5*h*(a(tn,ys[i]) + a(tn+h,ys[i+1]))
        
    return ts,ys,vs

ts,ys,vs = vel_verlet(a,0,15,1,0,0.01)

plt.plot(ts,ys)
plt.plot(ts,vs)
```

## problem 2  

A projectile of mass m = 2 kg is thrown with a speed of 10 ms−1 at an angle 60◦
w.r.t. the horizontal plane. The acceleration due to gravity is 9.8 ms2. The mass experiences a
velocity-dependent drag force with coefficients γ = 2 kg.s−1.

```python
def g(t,y,gamma,mass):
    return np.asarray([y[2] ,y[3], -(gamma/mass)*y[2],-((gamma/mass)*y[3]) -9.8])


gamma = 2
mass = 2

def euler(f,t_i,t_f,init_con,h):
    ts = np.arange(t_i,t_f,h)
    ys = np.zeros((len(ts),len(init_con)))
    
    ys[0] = init_con
    
    for i,tn in enumerate(ts[:-1]):
         ys[i+1] = ys[i] + h*g(tn,ys[i],2,2)
        
    return ts,ys
cond = np.asarray([0,0,10*np.cos(np.radians(60)),10*np.sin(np.radians(60))])
ts ,ys = euler(g,0,1.44,cond,0.01)

plt.plot(ys[:,0],ys[:,1])






def g(y,t,gamma,mass):
    return np.asarray([y[2] ,y[3], -(gamma/mass)*y[2],-((gamma/mass)*y[3]) -9.8])


def rk4(f,t_i,t_f,y_in,h):
    ts = np.arange(t_i,t_f,h)
    ys = np.zeros((len(ts),len(y_in)))
    
    ys[0] = y_in
    
    for i,tn in enumerate(ts[:-1]):
        k1 = h*f(ys[i],tn,2,2)
        k2 = h*f(ys[i] + k1/2 , tn + h/2 ,2,2)
        k3 = h*f(ys[i] + k2/2 , tn + h/2 ,2,2)
        k4 = h*f(ys[i] + k3 , tn + h,2,2)
        ys[i+1] = ys[i] + (k1 +2*k2 + 2*k3 + k4)/6
        
        
    return ts,ys


ts,ys = rk4(g,0,1.44,cond,0.01)

plt.plot(ys[:,0],ys[:,1])






```

## Velocity verlet for same problem 

```python

def ax(t,vx,gamma,mass):
    return -(gamma/mass)*vx

def ay(t,vy,gamma,mass,g):
    return ((-gamma/mass)*vy) - g



def vel_verlet(f,g,t_i,t_f,cond,h):
    
    ts = np.arange(t_i,t_f,h)
    xs = np.zeros(len(ts))
    ys = np.zeros(len(ts))
    vxs = np.zeros(len(ts))
    vys = np.zeros(len(ts))
    
    xs[0],ys[0],vxs[0],vys[0] = cond
    

    for i,tn in enumerate(ts[:-1]):
        xs[i+1] = xs[i] + vxs[i]*h + 0.5*f(tn,vxs[i],2,2)*h**2
        ys[i+1] = ys[i] + vys[i]*h + 0.5*g(tn,vys[i],2,2,9.8)*h**2
        vxs[i+1] = (2*mass*vxs[i] - h*gamma*vxs[i])/(2*mass + h*gamma)
        vys[i+1] = (2*mass*vys[i] - h*gamma*vys[i] -2*mass*h*9.8)/(2*mass + h*gamma)
        
    return ts,xs,ys,vxs,vys

ts,xs,ys,vxs,vys = vel_verlet(ax,ay,0,1.44,cond,0.01)

plt.plot(xs,ys)
```
