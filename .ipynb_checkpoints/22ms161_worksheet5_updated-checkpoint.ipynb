```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

## Definitions 

```python
def verlet(f,x,yn,ynm1,h):
    return 2*yn - ynm1 + f(x,yn)*(h**2)

def vel_verlet(f,x,yn,vn,h):
    ynp1 = yn + vn*h + f(x,yn)*(h**2)/2
    vnp1 = vn + (f(x,yn) + f(x+h,ynp1))*h*0.5
    return ynp1,vnp1

def euler(f, x, y, h):
    
    return y + h * f(x, y)


def rk4(f, x, y, h):
    
    k1 = h * f(x, y)
    k2 = h * f(x + h / 2, y + k1 / 2)
    k3 = h * f(x + h / 2, y + k2 / 2)
    k4 = h * f(x + h, y + k3)
    return y + (k1 + 2 * k2 + 2 * k3 + k4) / 6







def call_verlet(method,f,y0,v0,N,xs,h):
    
    ys = np.zeros(N)
    ys[0] = y0        # assume 1st value is initial value given 
    
    x0  = xs[0]      # first term of the interval of x
    
    ys[1] = y0 + v0*h + 0.5*f(x0,y0)*(h**2)   # explicit calculation of y1
    
    for i in range(2,N): # we need values from y2 now 
        x = xs[i]   # assigning values 
        ys[i]  = method(f,x,ys[i-1],ys[i-2],h)    # for 1st loop we gave y0 and y1 explicitly
        
    return ys



def call_method(method,f,y0,v0,N,xs,h):
    
    ys = np.zeros(N)
    vs = np.zeros(N)  # initializing the values just like dummy values
    
    ys[0] = y0
    vs[0] = v0  # assigning the initial values to the arrays
    
    for i in range(1,N):   # loop to get all the values based on the formula of the method
        x = xs[i]          # just assigning the values of x for calculation 
        
        ys[i],vs[i] = method(f,x,ys[i-1],vs[i-1],h)
        
    return ys,vs


        
def call_rk4(rk4,f,y_in,N,xs,h):
    
    y = y_in
    
    ys = np.zeros((N,len(y_in)),dtype = np.float64)
    
    for i in range(N):
        x= xs[i]
        ys[i,:] = y
        
        y = rk4(f,x,y,h)
    return ys

        
def call_euler(euler,f,y_in,N,xs,h):
    
    y = y_in
    
    ys = np.zeros((N,len(y_in)),dtype = np.float64)
    
    for i in range(N):
        x= xs[i]
        ys[i,:] = y
        
        y = euler(f,x,y,h)
    return ys
    
```

### Question 1


```python
# functions dy/dt = p
    #       dp/dt = -4*pi**2 y
    
def a(t,y):
    return -4*(np.pi)**2 * y
def g(t,y):
    return np.asarray([y[1] , -4*(np.pi)**2 * y[0]  ])  
    
ti = 0
tf = 15

h = 0.01

N =  int((abs(tf-ti)/h))




xs = np.linspace(ti,tf,N)    # so that last term is also included



y0 = 1                   # initial values
v0  = 0
                          

ys_verlet = call_verlet(verlet,a,y0,v0,N,xs,h)

ys_vv , vs_vv = call_method(vel_verlet,a,y0,v0,N,xs,h)


y_in = np.asarray([1.0,0.0])
ys_rk4 = call_rk4(rk4,g,y_in,N,xs,h)



fig,ax = plt.subplots(nrows = 3,ncols = 1 ,figsize=(9,9) )


# verlet 
ax[0].plot(xs,ys_verlet,label="velvet")





# velocity verlet
ax[1].plot(xs,ys_vv,label = "vel_velvet pos")
ax[1].plot(xs,vs_vv,label = "vel_velvet velocity")




# range kutta
ax[2].plot(xs,ys_rk4[:,0],label="rk4 pos")
ax[2].plot(xs,ys_rk4[:,1],label = "rk4 vel")



# exact

ax[0].plot(xs,np.cos(2*np.pi*xs),c= 'r',alpha = 0.5,label = "exact")
ax[0].legend()

ax[1].plot(xs,np.cos(2*np.pi*xs),c= 'r',alpha = 0.5,label = "exact")
ax[1].plot(xs,-2*np.pi*np.sin(2*np.pi*xs),c= 'g',alpha = 0.5,label = "exact velocity")
ax[1].legend()

ax[2].plot(xs,np.cos(2*np.pi*xs),c= 'r',alpha = 0.5,label = "exact")
ax[2].plot(xs,-2*np.pi*np.sin(2*np.pi*xs),c= 'g',alpha = 0.5,label = "exact velocity")
ax[2].legend()



ax[0].set_title("velvet")
ax[1].set_title("velocity velvet")
ax[2].set_title("rk4")

plt.tight_layout()
```

```python
error_vel = np.zeros(len(hs))
error_velocity_vel = np.zeros(len(hs))
error_rk4 = np.zeros(len(hs))
for i, h in enumerate(hs):
    xs = np.arange(0.0, 15.0 , h)
    N = len(xs)
    ys_v = call_verlet(verlet,a,y0,v0,N,xs,h)
    ys_vv , vs_vv = call_method(vel_verlet,a,y0,v0,N,xs,h)
    ys_rk4 = call_rk4(rk4,g,y_in,N,xs,h)
    
    
    error_vel[i] = abs(ys_v[-1] - np.cos(2*np.pi*xs[-1]))
    error_velocity_vel[i] = abs(ys_vv[-1] - np.cos(2*np.pi*xs[-1]))
    error_rk4[i] = abs(ys_rk4[-1,0] -  np.cos(2*np.pi*xs[-1]))

print("for velret errors:",error_vel)
print("\n")
print("for velocity errors:",error_velocity_vel)   
print("\n")
print("for rk4 errors",error_rk4)
```

```python
x1 = np.log10(hs)

def line(x,m,c):
    return m*x+c

ye1 = np.log10(error_vel)
ye2 = np.log10(error_velocity_vel)
ye3 = np.log10(error_rk4)

fig,ax = plt.subplots(nrows = 3,ncols = 1,figsize=(9,9))

ax[0].scatter(x1,ye1)
par0,parcov0 = curve_fit(line,x1,ye1)
ax[0].plot(x1,line(x1,*par0))
ax[0].annotate(f'm = {par0[0]}',xy = (-4,-6))
ax[0].set_title("velrvet")

ax[1].scatter(x1,ye2)
par1,parcov1 = curve_fit(line,x1,ye2)
ax[1].plot(x1,line(x1,*par1))
ax[1].annotate(f'm = {par1[0]}',xy = (-4,-6))
ax[1].set_title("vel_velret")


ax[2].scatter(x1,ye3)
par2,parcov2 = curve_fit(line,x1,ye3)
ax[2].plot(x1,line(x1,*par2))
ax[2].annotate(f'm = {par2[0]}',xy = (-4,-8))
ax[2].set_title("rk4")


plt.tight_layout()
```

##  Question 2



$$ F = F_{drag} + mg $$
$$  F_{drag} = -\gamma v $$

taking component wise $$ \dfrac{dv_x}{dt} = -\dfrac{\gamma}{2} v_x $$
$$  \dfrac{dv_y}{dt} = -\dfrac{\gamma}{2}v_y - 9.8  $$


$$ \dfrac{dx}{dt} = v_x   $$
$$ \dfrac{dy}{dt} = v_y   $$

$$  v_{y0} = 10sin(60) = 8.66 $$
$$  v_{x0} = 10cos(60) = 5 $$
$$  time of flight  \sim 1.45 sec $$

## velocity verlet for same problem  

```python
gamma = 2
mass = 2

cond = np.asarray([0,0,10*np.cos(np.radians(60)),10*np.sin(np.radians(60))])   # initial conditions

def ax(t,vx,gamma,mass):
    return -(gamma/mass)*vx     # seperate functions for x and y 

def ay(t,vy,gamma,mass,g):
    return ((-gamma/mass)*vy) - 9.8



def vel_verlet(f,g,t_i,t_f,cond,gamma,h):
    
    ts = np.arange(t_i,t_f,h)
    xs = np.zeros(len(ts))
    ys = np.zeros(len(ts))
    vxs = np.zeros(len(ts))
    vys = np.zeros(len(ts))
    
    xs[0],ys[0],vxs[0],vys[0] = cond
    

    for i,tn in enumerate(ts[:-1]):
        xs[i+1] = xs[i] + vxs[i]*h + 0.5*f(tn,vxs[i],2,2)*h**2
        ys[i+1] = ys[i] + vys[i]*h + 0.5*g(tn,vys[i],2,2,9.8)*h**2
        vxs[i+1] = (2*mass*vxs[i] - h*gamma*vxs[i])/(2*mass + h*gamma)    # did some modifications to vn+1 based on ODE
        vys[i+1] = (2*mass*vys[i] - h*gamma*vys[i] -2*mass*h*9.8)/(2*mass + h*gamma)    # explicit calculation of vn+1
        
        
    return ts,xs,ys,vxs,vys

ts,xs,ys,vxs,vys = vel_verlet(ax,ay,0,2,cond,gamma,0.01)
ys_positive  = ys[ys>0] 
plt.plot(xs[:len(ys_positive)],ys_positive,label="$\gamma = 2$")
plt.xlabel("x")
plt.ylabel("y")
plt.title("trajectory")
plt.legend()
plt.tight_layout()
plt.show()


print(np.array([xs,ys,vxs,vys]))
```

```python
def ax(t,vx,gamma,mass):
    return -(gamma/mass)*vx

def ay(t,vy,gamma,mass,g):
    return ((-gamma/mass)*vy) - g

for gamma in np.arange(0,11,1):
    def vel_verlet(f,g,t_i,t_f,cond,gamma,h):

        ts = np.arange(t_i,t_f,h)
        xs = np.zeros(len(ts))
        ys = np.zeros(len(ts))
        vxs = np.zeros(len(ts))
        vys = np.zeros(len(ts))

        xs[0],ys[0],vxs[0],vys[0] = cond


        for i,tn in enumerate(ts[:-1]):
            xs[i+1] = xs[i] + vxs[i]*h + 0.5*f(tn,vxs[i],gamma,mass)*h**2
            ys[i+1] = ys[i] + vys[i]*h + 0.5*g(tn,vys[i],gamma,mass,9.8)*h**2
            vxs[i+1] = (2*mass*vxs[i] - h*gamma*vxs[i])/(2*mass + h*gamma)           # did some modifications to vn+1 based on ODE
            vys[i+1] = (2*mass*vys[i] - h*gamma*vys[i] -2*mass*h*9.8)/(2*mass + h*gamma)        # explicit calculateion of vn+1.

        return ts,xs,ys,vxs,vys


    ts,xs,ys,vxs,vys = vel_verlet(ax,ay,0,2,cond,gamma,0.01)  # arbitrary time but we take only upto touching the ground
    ys_positive  = ys[ys>0] 
    plt.plot(xs[:len(ys_positive)],ys_positive,label=f"$\gamma = {gamma}$")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("trajectory")
    plt.legend()
    plt.xlim(-0.7,12)
    plt.tight_layout()
plt.show()    
```
